```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def poly_least_squares(t, y, deg):
    m = len(t)
    y = np.array(y).reshape(m,1)
    A = np.ones( (m,1) )
    
    t_row = t.reshape( (m,1) )
    row = np.ones( (m, 1) )
    for i in range(deg):
        row *= t_row
        A = np.concatenate( (A, row), axis=1)
    return np.linalg.lstsq(A, y,rcond=None)
```

```python
m = 100

'''
t = np.array(range(0,m))
y = [1,2,3]*((m)//3 + 1)
y = y[:m]
'''

x_min = 0
x_max = 2

t = np.linspace(x_min, x_max, m)
y = np.random.rand(m) + 5*t - 2*t**2 + .3*t**3 + .1*t**4 -.07*t**7
print(len(t))
print(len(y))



```

```python
plt.plot(t,y,'bo')


t1 = np.linspace(x_min, x_max, 1000)
for deg in range(2,7):
    p = np.poly1d( poly_least_squares(t,y,deg)[0].reshape(deg+1)[::-1] )
    plt.plot(t1, np.polyval(p,t1))
```
